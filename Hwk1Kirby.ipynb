```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y, X = np.mgrid[-5:5:100j, -5:5:100j]
U = X
V = -Y

plt.figure(num=1, figsize=(5,5))
plt.streamplot(X, Y, U, V, color=U, linewidth=1)
plt.title("a) Psi = Axy")
```

```python
Y, X = np.mgrid[-5:5:100j, -5:5:100j]
U = Y
V = 0*X

plt.figure(num=1, figsize=(5,5))
plt.streamplot(X, Y, U, V, color=U, linewidth=1)
plt.title("a) Psi = Axy")
```

```python
Y, X = np.mgrid[-5:5:100j, -5:5:100j]
U = Y/(X**2+Y**2)
V = -X/(X**2+Y**2)

plt.figure(num=1, figsize=(5,5))
plt.streamplot(X, Y, U, V, color=U, linewidth=1)
plt.title("a) Psi = C*ln(x^2+y^2)")
```

```python
Y, X = np.mgrid[-5:5:100j, -5:5:100j]
U = -2*Y
V = 2*X

plt.figure(num=1, figsize=(5,5))
plt.streamplot(X, Y, U, V, color=U, linewidth=1)
plt.title("a) Psi = Axy")
```
